```python
import numpy as np
import pandas as pd
import ast
```

```python
anime = pd.read_csv("animes.csv")
```

```python
anime.head()
```

```python
anime = anime[['uid','title','genre','synopsis']]
anime = anime.map(lambda x: x.lower() if isinstance(x, str) else x)
anime['genre'] = anime['genre'].apply(lambda x: ast.literal_eval(x))
```

```python
# (anime == 'UNKNOWN').sum()
# anime = anime.replace('UNKNOWN', pd.NA)
# (anime['synopsis'] == 'No description available for this anime.').sum()
# anime = anime.replace('No description available for this anime.', pd.NA)
# anime.isna().sum()
anime.isna().mean()*100
anime.dropna(inplace=True)
```

```python
anime = anime.reset_index(drop=True)
anime
```

```python

anime['synopsis'] = anime['synopsis'].apply(lambda x:x.split(' '))

```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python

anime['synopsis'] = anime['synopsis'].apply(collapse)
anime
```

```python
anime['tags'] = anime['genre']+ anime['synopsis']
new_anime = anime.drop(columns=['genre','synopsis'])
```

```python
new_anime['tags'] = new_anime['tags'].apply(lambda x: " ".join(x))
new_anime['tags']

```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_anime['tags'] = new_anime['tags'].apply(stem)
```

```python
new_anime
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6000,stop_words='english')
```

```python
vector = cv.fit_transform(new_anime['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
#new_anime[new_anime['title']== 'KImetsu no yaiba']
new_anime.sample(10)
```

```python
def recommend(anime):
  ind =  new_anime[new_anime['title'] == anime].index[0]
  print(ind)
  distances = sorted(list(enumerate(similarity[ind])),reverse=True,key = lambda x: x[1])
  for i in distances[1:10]:
    print(new_anime.iloc[i[0]]['title'])
```

```python
recommend('haikyuu!! second season')
```
